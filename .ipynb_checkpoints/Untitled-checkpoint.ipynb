```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
from scipy.fft import fft, fftfreq # Pour la FFT si vous allez plus loin

# --- 1. Charger les données labellisées ---
df = pd.read_csv('C:/Users/Kival/Desktop/fichier_PFA/sensor_data_labeled.csv')
df['timestamp_rpi'] = pd.to_datetime(df['timestamp_rpi'])
df = df.set_index('timestamp_rpi') # Indexer par le timestamp pour faciliter le fenêtrage

# --- 2. Gérer les valeurs manquantes (ex: -1.0 pour DHT) ---
# Remplacer les -1.0 par NaN pour les colonnes pertinentes
df['temperature_c'] = df['temperature_c'].replace(-1.0, np.nan)
df['humidity_percent'] = df['humidity_percent'].replace(-1.0, np.nan)
# Pour les NaN, vous pouvez les remplir (forward fill, back fill, moyenne, ou suppression)
df = df.ffill().bfill() # Remplir avec la dernière/première valeur connue

# --- 3. Ingénierie des Caractéristiques par Fenêtre ---
window_size_seconds = 1 # Taille de la fenêtre en secondes (ex: 1s, 2s, etc.)
# 'S' pour secondes dans resample. 'T' pour minutes, 'H' pour heures etc.
# 'origin='start_day'' est utile pour aligner les fenêtres au début de la journée.

# Liste pour stocker les caractéristiques de chaque fenêtre
features_list = []

# Itérer sur les fenêtres de temps
# 'offset' permet de décaler le début des fenêtres si nécessaire
# 'closed='right'' inclut la fin de la fenêtre. 'label='right'' met le timestamp à la fin de la fenêtre.
for window_start, window_df in df.resample(f'{window_size_seconds}s', origin='start_day', closed='right', label='right'):
    if not window_df.empty:
        # Extraire le label de la fenêtre (le plus fréquent)
        window_label = window_df['fault_type'].mode()[0] # mode() retourne une Series, [0] pour la première valeur

        # Calculer les caractéristiques pour cette fenêtre
        features = {
            'window_start': window_start - pd.Timedelta(seconds=window_size_seconds), # Début réel de la fenêtre
            'window_end': window_start, # Fin de la fenêtre (qui est le label de temps de resample)
            'label': window_label,

            # Caractéristiques de Température
            'temp_mean': window_df['temperature_c'].mean(),
            'temp_std': window_df['temperature_c'].std(),
            'temp_min': window_df['temperature_c'].min(),
            'temp_max': window_df['temperature_c'].max(),

            # Caractéristiques d'Humidité
            'hum_mean': window_df['humidity_percent'].mean(),
            'hum_std': window_df['humidity_percent'].std(),
            'hum_min': window_df['humidity_percent'].min(),
            'hum_max': window_df['humidity_percent'].max(),

            # Caractéristiques du Bruit (simulé)
            'noise_mean': window_df['noise_level_simulated'].mean(),
            'noise_std': window_df['noise_level_simulated'].std(),

            # Caractéristiques de Vibration (Accélération X, Y, Z)
            # RMS
            'accel_x_rms': np.sqrt(np.mean(window_df['accel_x_g']**2)),
            'accel_y_rms': np.sqrt(np.mean(window_df['accel_y_g']**2)),
            'accel_z_rms': np.sqrt(np.mean(window_df['accel_z_g']**2)),
            
            # Écart-type
            'accel_x_std': window_df['accel_x_g'].std(),
            'accel_y_std': window_df['accel_y_g'].std(),
            'accel_z_std': window_df['accel_z_g'].std(),

            # Pic-à-Pic (Peak-to-Peak)
            'accel_x_ptp': window_df['accel_x_g'].max() - window_df['accel_x_g'].min(),
            'accel_y_ptp': window_df['accel_y_g'].max() - window_df['accel_y_g'].min(),
            'accel_z_ptp': window_df['accel_z_g'].max() - window_df['accel_z_g'].min(),

            # Skewness
            'accel_x_skew': skew(window_df['accel_x_g']),
            'accel_y_skew': skew(window_df['accel_y_g']),
            'accel_z_skew': skew(window_df['accel_z_g']),

            # Kurtosis
            'accel_x_kurt': kurtosis(window_df['accel_x_g']),
            'accel_y_kurt': kurtosis(window_df['accel_y_g']),
            'accel_z_kurt': kurtosis(window_df['accel_z_g']),
        }
        features_list.append(features)

# Créer un nouveau DataFrame avec les caractéristiques extraites
features_df = pd.DataFrame(features_list)
print("DataFrame des caractéristiques créé avec succès.")
print(features_df.head())
print(f"Nombre de fenêtres (échantillons pour l'IA) : {len(features_df)}")

# Sauvegarder ce DataFrame si vous voulez le réutiliser sans recalculer
features_df.to_csv('features_for_training1.csv', index=False)
```

```python
import pandas as pd

# Charger les caractéristiques extraites + labels depuis un fichier CSV
features_df = pd.read_csv('C:/Users/Kival/Desktop/fichier_PFA/features_for_training1.csv')  

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Séparer les caractéristiques (X) des labels (y)
X = features_df.drop(['window_start', 'window_end', 'label'], axis=1)
y = features_df['label']

# Encoder les labels textuels en numériques
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y) # Ex: 'Normal' -> 0, 'Unbalance' -> 1, 'Friction' -> 2

# Afficher les correspondances
print("Correspondance des labels numériques :")
for i, label in enumerate(label_encoder.classes_):
    print(f"{label} -> {i}")

# Diviser les données en ensembles d'entraînement et de test
# random_state assure la reproductibilité de la division
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

print(f"\nTaille de l'ensemble d'entraînement (X_train) : {X_train.shape}")
print(f"Taille de l'ensemble de test (X_test) : {X_test.shape}")
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Créer une instance du classifieur XGBoost
# Vous pouvez ajuster les hyperparamètres comme n_estimators, learning_rate, max_depth
model_xgb = xgb.XGBClassifier(objective='multi:softmax',  # Pour classification multi-classes
                              num_class=len(label_encoder.classes_), # Nombre de classes uniques
                              eval_metric='mlogloss', # Métrique d'évaluation
                              #use_label_encoder=False, # Recommandé pour les versions récentes
                              n_estimators=100,       # Nombre d'arbres
                              learning_rate=0.1,      # Taux d'apprentissage
                              max_depth=5,            # Profondeur max des arbres
                              random_state=42)

# Entraîner le modèle
print("\nEntraînement du modèle XGBoost...")
model_xgb.fit(X_train, y_train)
print("Entraînement terminé.")

# --- Évaluation du modèle ---
y_pred_xgb = model_xgb.predict(X_test)

print("\n--- Rapports d'évaluation XGBoost ---")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("\nClassification Report:\n", classification_report(y_test, y_pred_xgb, target_names=label_encoder.classes_))

# Matrice de confusion
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title('Matrice de Confusion (XGBoost)')
plt.xlabel('Prédit')
plt.ylabel('Réel')
plt.show()

# --- Visualisation de l'importance des caractéristiques ---
# Ceci est très utile pour comprendre ce que le modèle a appris
plt.figure(figsize=(12, 6))
xgb.plot_importance(model_xgb, max_num_features=20) # Affiche les 20 caractéristiques les plus importantes
plt.title('Importance des Caractéristiques (XGBoost)')
plt.show()
```
